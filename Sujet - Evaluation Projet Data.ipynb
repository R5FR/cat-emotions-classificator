# Sujet d'evaluation - Projet Data (travail en groupe)

## Contexte
Vous travaillez en groupe (2 a 4 etudiants) pour realiser un **projet Data complet** a partir d'un jeu de donnees de votre choix (ou propose par l'enseignant).

L'objectif est de conduire un projet de Machine Learning de bout en bout, avec une demarche rigoureuse, argumentee et explicable.

## Objectif pedagogique

A la fin de ce projet, vous devez etre capables de :
- formuler une problematique Data claire ;
- preparer et nettoyer un jeu de donnees reel ;
- realiser une analyse exploratoire (EDA) pertinente ;
- tester plusieurs modelisations ;
- comparer les resultats et justifier vos choix ;
- expliquer votre demarche de facon compréhensible.

## Etapes obligatoires du projet

### 1) Definition du probleme
- Presenter le contexte metier ou applicatif.
- Definir la variable cible (ou l'objectif d'analyse si non supervise).
- Enoncer les hypotheses de depart.

### 2) Collecte et comprehension des donnees
- Identifier la source des donnees.
- Decrire les variables disponibles (type, role, unite, etc.).
- Evaluer rapidement la qualite des donnees.

### 3) Nettoyage et pretraitement
- Gestion des valeurs manquantes.
- Detection et traitement des doublons et valeurs aberrantes.
- Encodage des variables categorielles.
- Normalisation/standardisation si necessaire.
- Creation eventuelle de nouvelles variables (feature engineering).

### 4) EDA (analyse exploratoire des donnees)
- Statistiques descriptives.
- Visualisations pertinentes (distributions, correlations, comparaisons, etc.).
- Identification de tendances, biais et limites du dataset.

### 5) Modelisation
- Tester au moins **2 modeles differents**.
- Justifier le choix des metriques d'evaluation.
- Mettre en place un split train/test (et validation croisee si pertinent).
- Comparer les performances de facon objective.

### 6) Interpretation et explicabilite
- Expliquer les resultats obtenus en lien avec le probleme initial.
- Identifier les variables importantes.
- Presenter les limites, risques de biais et pistes d'amelioration.

### 7) Conclusion
- Faire une synthese des enseignements cles.
- Proposer des recommandations concretes ou suites possibles du projet.

## Livrables attendus

1. **Notebook Jupyter** propre et execute, contenant toutes les etapes du projet.
2. **Rapport final** au choix :
   - soit en **PDF**,
   - soit sous forme d'**application web** (ex: Streamlit, Dash, Gradio), documentee.
3. (Optionnel) Depot GitHub avec code, README et instructions d'execution.

## Grille d'evaluation (sur 20)

- **Qualite de l'analyse** (8 points)
  - Pertinence de la problematique
  - Qualite de l'EDA
  - Rigueur du nettoyage/pretraitement

- **Qualite de la modelisation** (4 points)
  - Diversite et pertinence des modeles testes
  - Choix des metriques et interpretation des performances

- **Explicabilite et argumentation** (6 points)
  - Capacite a expliquer les choix methodologiques
  - Capacite a interpreter les resultats clairement
  - Discussion des limites et biais

- **Clarte de la restitution** (2 points)
  - Structure du notebook/rapport
  - Qualite de la communication ecrite et visuelle

## Conseils de realisation

- Soyez explicites sur vos choix (et pas uniquement sur les resultats).
- Privilegiez des visualisations lisibles et commentees.
- Assurez la reproductibilite (imports, seed, pipeline clair).
- Structurez votre notebook pour qu'une personne externe puisse suivre votre logique.
